```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('/content/drive/MyDrive/dice.png')

R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
print(imgGray.shape)
plt.imshow(imgGray, cmap='gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import copy

imgGray = cv2.imread('/content/drive/MyDrive/dice.png', 0)
plt.title('Original image')  
plt.imshow(imgGray, cmap='gray')
plt.show()

plt.title('Negative Tranformation')  
plt.imshow(255 - imgGray, cmap='gray')
plt.show()

uinput = input("Enter  constant value")
try:
  plt.title("Adding constant")
  plt.imshow(abs(  imgGray + int(uinput)%255), cmap='gray')
  plt.show()
except:
  print("Invalid input")

uinput = input("Enter  threshold value")
if(uinput.isdigit()):
  plt.title("Binary image")
  imgbinary = imgGray.copy()
  for i in range(len(imgbinary)):
    for j in range(len(imgbinary[0])):
        if(imgbinary[i][j] > int(uinput)):
            imgbinary[i][j] = 1
        else:
          imgbinary[i][j] = 0
  plt.imshow(imgbinary, cmap='gray')
  plt.show()
else:
  print("Invalid input")


imgLog = imgGray.copy()
c = 255 / (np.log (1 + 255))
for i in range(len(imgLog)):
    for j in range(len(imgLog[0])):
        imgLog[i][j] = c*np.log(1+imgLog[i][j])

plt.title('Log Transformation')  
plt.imshow(imgLog, cmap='gray')
plt.show()


uinput = input("Enter  gamma value")
try:
  imgexp = imgGray.copy()
  c = 255 / (np.log (1 + 255))
  for i in range(len(imgexp)):
      for j in range(len(imgexp[0])):
          imgexp[i][j] = c*pow(imgexp[i][j], float(uinput))
  plt.title("Exponential Tranformation")
  plt.imshow(imgexp, cmap='gray')
  plt.show()
except:
  print("Invalid input")






```
